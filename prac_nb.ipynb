```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/spam.csv',encoding='latin-1')
```

```python
df
```

```python
df.rename(columns={'v1':'Class_label','v2':'message'},inplace=True)
```

```python
df
```

```python
df.drop(['Unnamed: 2'],axis=1,inplace=True)
```

```python
df.drop(['Unnamed: 3', 'Unnamed: 4'],axis=1,inplace=True)
```

```python
df
```

```python
df.Class_label.value_counts()
```

```python
df['Class_label']=df['Class_label'].apply(lambda x: 1  if x=="spam" else 0 )
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df['message'],df['Class_label'],test_size=0.3,random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer(lowercase=True,stop_words='english')
```

```python
X_train_transformed=vectorizer.fit_transform(X_train)
X_test_transformed=vectorizer.transform(X_test)
```

```python
X_train_transformed.toarray()
```

```python
vectorizer.get_feature_names_out()
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
classifier.fit(X_train_transformed,y_train)
```

```python
y_test_predicted=classifier.predict(X_test_transformed)
```

```python
y_test_predicted
```

```python
y_test.count()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('Accuracy Score:',accuracy_score(y_test,y_test_predicted))
```

```python
results=confusion_matrix(y_test,y_test_predicted)
results
```

```python
print(classification_report(y_test,y_test_predicted))
```
